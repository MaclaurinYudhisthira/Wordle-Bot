# Counting the no of occernce of letter in _`whole df`_

```python
adf=gdf
```

```python
def count_letter_in_df(df,df1):
    for letter in "abcdefghijklmnopqrstuvwxyz":
        df[letter]=pd.Series([word.count(letter) for word in df1['words']])
```

# Calculating the probability of occernce of letter on a specific position _`whole df`_ `P2`

```python
df1=pd.DataFrame()
```

```python
def count_letter_at_position_in_df(df,df1,letter):
    for i in range(5):
        df[f"{i+1}"]=pd.Series([1 if letter==word[i] else 0 for word in df1['words']])
```

```python
for letter in "abcdefghijklmnopqrstuvwxyz":
    df=pd.DataFrame()
    count_letter_at_position_in_df(df,adf,letter)
    df1[letter]=df.sum()
```

# Sorting the df by calculating P1 based Weightage

```python
probability_of_occurance
```

```python
probability_of_occurance['a']
```

```python
probability_of_occurance['b']
```

```python
df=adf.copy()
df
```

```python
def Calculate_P1_Based_Weightage(row):
    weight=0
    for letter in row['words']:
        weight+=probability_of_occurance[letter]
    return weight  
```

```python
df['weightage']=df.apply(lambda row: Calculate_P1_Based_Weightage(row), axis=1)
```

```python
df=df.sort_values(by=['weightage'],ascending=False)
```

```python
df.head()
```

```python
for letter in "abcdefghijklmnopqrstuvwxyz":
    df[letter]=[1 if letter in word else 0 for word in df['words']]
```
